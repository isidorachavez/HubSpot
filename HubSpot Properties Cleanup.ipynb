```python
"""
author: Isidora Chavez

Description: This code is to cleanup HubSpot properties exported file to see 
the options provided in dopdown or checklist fields.

"""

import pandas as pd

#put csv file name
df = pd.read_csv("contact.csv") 
```

```python
#makes N/A in Options column as blank
df['Options'] = df['Options'].fillna("")
```

```python
#splitting Options column
# placeholder list
placeholder_lst = []

# looping over each email address
for index, col in df.iterrows():
    
    # splitting options at 'label":"'
    split_options = df.loc[index, 'Options'].split(sep = 'label":"')
    # splitting further after '"'
    split_options = df.loc[index, 'Options'].split(sep = '"')
    # appending placeholder_lst with the results
    placeholder_lst.append(split_options)
    

# converting placeholder_lst into a DataFrame 
options_df = pd.DataFrame(placeholder_lst)
# slice to only get column 3 onwards - HubSpot's format shows this pattern
options_df = options_df.iloc[:,3:] 
```

```python
#keeps only every 20th column - HubSpot's format shows this pattern
options_new = options_df.iloc[:, ::20]
```

```python
#This makes the first dataframes 'Name' column into a new dataframe
fields = pd.DataFrame(df['Name'])

#These are the two dataframes we want
field_options = [fields , options_new] 

#Merging the two dataframes side by side, use axis = 1
field_options = pd.concat(field_options, axis=1)
```

```python
#You can type whatever name you want for the final csv file
field_options.to_csv("Contacts Options.csv")
```
